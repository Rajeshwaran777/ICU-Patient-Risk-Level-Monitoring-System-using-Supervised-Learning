## DATA VISUALIZATION AND DATA ANALYSIS

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('ICU.csv')

del df['ID']
del df['gendera']
del df['group']

df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(12, 7))
sns.set(style="whitegrid")  # Adding a white grid background

# Customize the countplot appearance
sns.countplot(x='outcome', data=df, palette='pastel')  # You can specify a custom color palette

# Adding labels and title
plt.xlabel('Outcome', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Outcomes', fontsize=15)

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)  # You can adjust the rotation angle as needed

# Adding count labels above the bars
ax = sns.countplot(x='outcome', data=df, palette='pastel')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=12, color='black')

plt.show()
```

```python
# Create a figure with two subplots side by side
plt.figure(figsize=(15, 5))

# Subplot 1: Histogram for 'SP O2'
plt.subplot(1, 2, 1)
plt.hist(df['SP O2'], color='red', bins=20)  # You can adjust the number of bins
plt.xlabel('SP O2', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of SP O2', fontsize=15)

# Subplot 2: Histogram for 'Lymphocyte'
plt.subplot(1, 2, 2)
plt.hist(df['Lymphocyte'], color='blue', bins=20)  # You can adjust the number of bins
plt.xlabel('Lymphocyte', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Lymphocyte', fontsize=15)

# Ensure subplots don't overlap
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Specify the figure size and layout
fig, axes = plt.subplots(15, 4, figsize=(15, 55))
fig.subplots_adjust(hspace=0.5)

# Define a list of colors for the histograms
colors = ['red', 'blue', 'green', 'orange', 'purple', 'pink', 'brown', 'gray', 'cyan', 'magenta', 'yellow', 'lime', 'teal', 'navy', 'salmon']

# Loop through each column and create histograms with different colors
for i, col in enumerate(df.columns):
    ax = axes[i // 4, i % 4]
    color = colors[i % len(colors)]  # Cycle through the color list
    ax.hist(df[col], color=color, bins=20)  # You can adjust the number of bins
    ax.set_title(col, fontsize=12)
    ax.set_xlabel(col, fontsize=10)
    ax.set_ylabel('Frequency', fontsize=10)

# Hide any empty subplots
for i in range(len(df.columns), 15 * 4):
    fig.delaxes(axes[i // 4, i % 4])

plt.show()
```

```python
sns.violinplot(x='Urine output', y='EF', data=df, color='pink')
sns.swarmplot(x='Urine output', y='EF', data=df, color='blue', size=5)
plt.xlabel('Urine output')
plt.ylabel('EF')
plt.title('Violin Swarm Plot')
plt.show()
```

```python
def plot_pie(df, variable, title=None, fontsize=10, colors=None):
    dataframe_pie = df[variable].value_counts()
    
    # Set a custom color palette (e.g., 'Set3' is one of the available color palettes)
    sns.set_palette(colors or 'Set3')
    
    ax = dataframe_pie.plot.pie(figsize=(9, 9), autopct='%1.2f%%', fontsize=fontsize)
    
    if title:
        ax.set_title(title + ' \n', fontsize=fontsize)
    
    plt.show()  # Display the pie chart
    return dataframe_pie

# Example usage with a custom color palette
custom_colors = ['grey', 'lightcoral', 'lightgreen', 'lightsalmon']
plot_pie(df, 'outcome', title='Outcome Distribution', fontsize=12, colors=custom_colors)
```
